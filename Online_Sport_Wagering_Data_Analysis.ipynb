```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Selected_Online_Sport_Wagering_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Convert 'Month Ending' to datetime
df['Month Ending'] = pd.to_datetime(df['Month Ending'])
df['Month Ending']
```

```python
# Summary statistics of numerical columns
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Visualize the distribution of numerical features
numerical_features = df.select_dtypes(include=['int64']).columns
df[numerical_features].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features', y=1.02)
plt.show()
```

```python
# Explore the correlation between features
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month Ending', y='Wagers', data=df, marker='o')
plt.title('Trends in Wagers Over Time')
plt.xlabel('Month Ending')
plt.ylabel('Wagers')
plt.show()
```

# Distribution of Categorical Features

```python
# Visualize the distribution of categorical features (e.g., Licensee, Fiscal Year)
categorical_features = df.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, data=df)
    plt.title(f'Distribution of {feature}')
    plt.xticks(rotation=45)
    plt.show()
```

# Trends in Total Gross Gaming Revenue and Payments Over Time

```python
# Visualize trends in Total Gross Gaming Revenue and Payments Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month Ending', y='Total Gross Gaming Revenue', data=df, marker='o', label='Total Gross Gaming Revenue')
sns.lineplot(x='Month Ending', y='Payment (7)', data=df, marker='o', label='Payments')
plt.title('Trends in Revenue and Payments Over Time')
plt.xlabel('Month Ending')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

# Payment Rate Analysis

```python
# Calculate and visualize the payment rate over time
df['Payment Rate'] = df['Payment (7)'] / df['Total Gross Gaming Revenue'] * 100

plt.figure(figsize=(12, 6))
sns.lineplot(x='Month Ending', y='Payment Rate', data=df, marker='o')
plt.title('Payment Rate Over Time')
plt.xlabel('Month Ending')
plt.ylabel('Payment Rate (%)')
plt.show()
```

# Explore Correlation with Pairplot

```python
# Pairplot to visualize relationships between numerical features
sns.pairplot(df[numerical_features])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```
